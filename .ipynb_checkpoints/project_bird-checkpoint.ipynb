```python
def df_process(df_data=df_data):
    df_data_processed = df_data.dropna(subset=['timestamp', 'location-long','ground-speed','location-lat','event-id'])
    dt = pd.to_datetime(df_data_processed[['timestamp']].stack(), format='%Y-%m-%d %H:%M:%S')
    df_data_processed['timestamp'] = dt.droplevel(level=1)
    df_data_processed = df_data_processed.sort_values(["individual-local-identifier","timestamp"]).set_index(["individual-local-identifier"])
    # add time diff
    df_data_processed = df_data_processed.rename(columns = {
        'location-lat':'Latitude', 'location-long':'Longitude',"eobs:temperature":"Temperature"})
    cols_to_care = ["timestamp","Latitude","Longitude","Temperature","ground-speed","heading"]
    return df_data_processed[cols_to_care]

def df_engineer(df_process):
    df_final = df_process.copy()
    df_final["delta_time"] = df_data_processed.groupby(level=0)["timestamp"].diff(1).shift(-1)
    df_final['delta_time(s)'] = df_final['delta_time'].apply(lambda x: x.total_seconds())
    df_final = df_final.dropna(subset=["delta_time(s)"]).set_index("timestamp", append=True)
    return df_final
df_data_processed = df_process(df_data); df_data_processed.head(2)
df = df_engineer(df_data_processed); df.head(2)
```

### Positions Encoding

```python
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.cluster import KMeans
import colorsys
```

```python
n_clusters = np.arange(30, 35, 5); l = {}
x_loc = df[["Latitude","Longitude"]][1:]
dt = df["delta_time(s)"]
dt_scaler = dt/(dt.max()-dt.min())
```

```python
for n in n_clusters:
    km = KMeans(n_clusters=n)
    km.fit(x_loc, np.e**(dt_scaler[1:]))
    l[n] = km.inertia_
print("Done...")
```

```python
fig, ax = plt.subplots(1,1)
x,y = zip(*l.items())
ax.plot(x,y)
ax.set_xlabel("N clusters"); ax.set_ylabel("Loss"); ax.set_title("Cluster Select by Loss")
fig.savefig(proj_path/"Weighted Cluster select by exponential loss (20-50).jpg")
```

```python
y_kmeans = km.predict(x_loc); n_c = len(np.unique(y_kmeans))
colours = cm.rainbow(np.linspace(0, 1, n_c))
for i_c, color in zip(range(n_c), colours):
    plt.scatter(x_loc.iloc[y_kmeans==i_c, 1], x_loc.iloc[y_kmeans==i_c, 0],marker=".",
                s=1, c=[color])
plt.scatter(km.cluster_centers_[:, 1], km.cluster_centers_[:, 0],
            s=20, c='black', label = 'Centroids')
plt.title('Clusters of location'); plt.xlabel('Latitude'); plt.ylabel('Longtitude')
plt.savefig(proj_path/"kmeans of locations (exp)")
```

fit(X, lengths=None)
Estimate model parameters.

An initialization step is performed before entering the EM algorithm. If you want to avoid this step for a subset of the parameters, pass proper init_params keyword argument to estimator’s constructor.

Parameters
X (array-like, shape (n_samples, n_features)) – Feature matrix of individual samples.

lengths (array-like of integers, shape (n_sequences, )) – Lengths of the individual sequences in X. The sum of these should be n_samples.

Returns
self (object) – Returns self.

```python
class bird_island_hmm:
    def __init__(self, feats, state):
        # (n, s_len) # (n, n_feat, s_len)
        self.states, self.emissions = self.prep_hmm_input(feats, state)
        self.n_state = len(np.unique(self.states))
        self.n_feat = self.emissions[0].shape[1]
        print("Number of unique state", self.n_state)
        print("Check input data len", len(self.states), len(self.emissions),
             "Shape", self.states[0].shape, self.emissions[0].shape)
        pass
    
    def fit_markov_chain(self, df):
        self.trans_probs = np.zeros((self.n_state, self.n_state))
        for s in self.states:
            for n in range(len(s[:-1])):
                self.trans_probs[s[n], s[n+1]] += 1
        self.emit_probs = df[feats+state].groupby(state[0]).mean()
        return self.trans_probs/np.sum(self.trans_probs, axis=0, keepdims=True), self.emit_probs
        
    def plot_gaussian_emit(self):
        #x-axis ranges from -3 and 3 with .001 steps
        x = np.arange(-3, 3, 0.001)
        #plot normal distribution with mean 0 and standard deviation 1
        plt.plot(x, norm.pdf(x, 0, 1))

    def prep_hmm_input(self, feats, state):
        self.states = []
        self.emissions = []
        for ID in birds_id:
            self.states.append(df.loc[ID]["island"].values.T)
            self.emissions.append(df.loc[ID][feats].values)
        max_len = max([len(s) for s in self.states])
        terminal_state = max([max(s) for s in self.states])
        # pad sequence to the same length with final stop
        for n, (s,e) in enumerate(zip(self.states, self.emissions)):
            self.states[n] = np.pad(s, ((0, max_len-len(s))), 'constant', constant_values=(0, terminal_state))
            self.emissions[n] =np.pad(e, ((0, max_len-e.shape[0]), (0,0)), 'mean')
        return self.states, self.emissions    
    
    def train(self, hmm_args, lim_n = 3):    # EM train
        '''
        lim_n : limit number of trajectory to reduce run time
        '''
        self.hmm = GaussianHMM(**hmm_args)
        self.hmm.startprob_ = np.ones(self.n_state)/self.n_state    # Assume 
        self.hmm.transmat_ = np.identity(self.n_state)/self.n_state   # Assume along diagonal online
        self.hmm.fit(X = np.concatenate(self.emissions), lengths=[len(self.emissions[0])]*len(self.emissions))
        return self.hmm
    
    def sample(self, length = 10000):
        X, Z = self.hmm.sample(length)
        return X, Z
```
